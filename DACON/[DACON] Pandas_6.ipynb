### Ch06 컬럼 연산
- 목표 : 컬럼 간에 사칙연산을 수행

```python
import pandas as pd

data = pd.read_csv('data/data03.csv')
data
```

# 산술 연산
- 기존에 가지고 있는 컬럼들을 이용하여 새로운 컬럼을 만들 수 있습니다.
  
    
- 판다스의 컬럼끼리는 산술연살자를 이용하여 연산이 가능합니다.

```python
# 컬럼끼리 덧셈
data['남자+여자'] =data['남자'] + data['여자']
data
# print(data['인구수'] == data['남자+여자'])
```

```python
# 컬럼끼리 뺄쎔
data['남자 - 여자'] =data['남자'] -  data['여자']
data
```

```python
# 컬럼끼리 차이의 절대값
data['남자 - 여자_절대값'] = data['남자 - 여자'].abs()

data
```

```python
# 컬럼 비율 
data['남자/여자'] = data['남자']/data['여자']
data
```

- 하나의 상수를 이용한 연산


```python
data['여자_100명당_남자수'] = data['남자/여자']*100
data
```

```python
data.to_csv('data/data06.csv', index = False)

data.to_csv('data/data06_pandas.csv', index = False)
```
